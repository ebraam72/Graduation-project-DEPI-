```python
df[['order-Day', 'order-Month', 'order-Year']] = df['Order Date'].str.split("/",expand=True)
```

```python
df[['ship-Day', 'ship-Month', 'ship-Year']] = df['Ship Date'].str.split("/",expand=True)
```

```python
df.head()
```

```python
df.loc[:, 'Postal Code']=df['Postal Code'].fillna(54010)
```

```python
df.isnull().sum()
```

```python
df['Ship-Quarter'] = df['ship-Month'].apply(lambda x: 'Q1' if int(x) in [1, 2, 3] 
                                  else 'Q2' if int(x) in [4, 5, 6] 
                                  else 'Q3' if int(x) in [7, 8, 9] 
                                  else 'Q4')
```

```python
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y", errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format="%d/%m/%Y", errors='coerce')
df['Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
```
